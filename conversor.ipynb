```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import text
```

```python
df = pd.read_csv("g14_Airports_Airlines.csv", encoding='latin1', sep = ';', parse_dates=['build_date','flight_date'], dayfirst=True)
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.insert(0, 'voo_id', range(1, len(df) + 1))
```

```python
df.head()
```

```python
engine = create_engine('sqlite:///Airport airlines - 2.db')
```

```python
#Terminal
```

```python
df_terminal = df.loc[:,['terminal_id','notes','airport_id']]
```

```python
df_terminal.head(3)
```

```python
df_terminal = df_terminal.drop_duplicates(subset='terminal_id', keep='first')
```

```python
df_terminal = df_terminal.loc[df_terminal['terminal_id'].notnull()]
```

```python
df_terminal.rename(columns={'terminal_id':'id'}, inplace=True)
```

```python
df_terminal.head(3)
```

```python
with engine.begin() as conn:
    conn.execute(text("DELETE FROM Terminal;"))
#ao usar o append na de baixo, se ja estiverem la dados estes nao sao apagados. por isso, ao adicionar la novos vai dar erro se forem iguais aos existentes (pk implica que so haja uma linha com esse valor)
#havia duas formas de resolver isto: 1 - ao inserir, se o dado ja existir na tabela, ignorar ou replace (mas isso ja seria muito complicado!) ou 2 (usada aqui) - apagar todos os dados da tabela antes de inserir novos
#como neste exercicio estamos perante a inicializacao de uma tabela a partir de dados de um ficheiro csv, faz sentido usar a segunda opcao
```

```python
df_terminal.to_sql('Terminal', con=engine, if_exists='append', index=False) 
#nao posso usar aqui if_exists='replace' pq assim estraga me as pk e fk
```
